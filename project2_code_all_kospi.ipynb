```python
non_adm_input = kospi_720df.drop(columns=['DATE', 'YYMM', 'FS_YEAR', 'FS_MONTH', 'FS_QTR',
       'MAIN_TERM', 'M111000', 'M122700', 'M131000', 'M221000', 'M222000',
       'M113000', 'M121500', 'M231000'])
non_adm_input
```

```python
adm_input = pd.read_excel("adm_corp2222.xlsx")
adm_input
```

```python
adm_input["CODE"] = "A" + adm_input["CODE"].astype('str').str.zfill(6)
adm_input
```

```python
data_input = pd.concat([non_adm_input, adm_input])
data_input
```

```python
data_input.isnull().sum()
```

```python
data_input["stability_1"].values
```

```python
input_df = data_input.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split


x = input_df.iloc[:,2:8]
y = input_df["decision"]


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=121)


dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
dtree.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = dtree.predict(X_test)
confusion_matrix(y_test.values, y_pred, labels=[0,1])
```
